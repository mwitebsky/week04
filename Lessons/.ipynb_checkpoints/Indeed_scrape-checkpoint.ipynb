## Scraping Job Postings

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import urllib2
from bs4 import BeautifulSoup
import time
pd.set_option('display.max_row', 1000)
```

```python
cities_states = [['Chicago', 'IL'], ['Washington', 'DC'], ['San Francisco', 'CA'], ['New York', 'NY'], 
                 ['Houston', 'TX'], ['Seattle', 'WA'], ['Boston', 'MA'], ['San Diego', 'CA'], 
                 ['Philadelphia', 'PA'], ['Atlanta', 'GA'], ['Los Angeles', 'CA'], ['St. Louis', 'MO'], 
                 ['Minneapolis', 'MN'], ['Denver', 'CO'], ['Dallas', 'TX'], ['Austin', 'TX'], 
                 ['Phoenix', 'AZ'], ['San Antonio', 'TX'], ['San Jose', 'CA'], ['Indianapolis', 'IN'],
                 ['Charlotte', 'NC'], ['Detroit', 'MI'], ['Nashville', 'TN'], ['Las Vegas', 'NV'],
                 ['Baltimore', 'MD'], ['Louisville', 'KY'], ['Milwaukee', 'WI'], ['Albuquerque', 'NM'],
                 ['Fresno', 'CA'], ['Sacramento', 'CA'], ['Kansas City', 'MO'], ['Raleigh', 'NC']]
miles = '25'
search = 'data+scientist'   #  use + instead of space
# job_type = ''    #  fulltime, parttime, contract, internship, temporary, commission
# experience = ''    #  entry_level, mid_level, senior_level
results = 100

def append_result(result):
    result = result.get_text().strip()
    if len(result) > 1:
        return result
    else:
        return None
    
data = {}   
counter = 0

### Removed because pulling more than 100 results per city was
### causing strange results -- many duplicate posts
# find how many results per city
# for i in range(len(cities_states)):
#     time.sleep(1)
#     url = 'http://www.indeed.com/jobs?as_and=' + search + \
#         '&as_phr=&as_any=&as_not=&as_ttl=&as_cmp=&jt=all&st=&salary=&radius=' \
#         + str(miles) + '&l=' + cities_states[i][0] + '%2C+' + cities_states[i][1] + \
#         '&fromage=any&limit=50&sort=date&start=&psf=advsrch'
#     url_data = urllib2.urlopen(url).read()
#     soup = BeautifulSoup(url_data)
#     results_lst = []    
#     for div in soup('div', {'id' : 'searchCount'}):
#         results_lst = div.text.split()
#         try:    # some searches return no results: assume 1000
#             results = results_lst[5]
#         except IndexError:
#             results = '1000'
#         results = min(int(results.replace(',','')), 1000)    # weird scrapes after 1000 results
###
###

# Advanced search on city radius of 25mi, 50 results per page
for i in range(len(cities_states)):
    for n in range(0, results-1, 50):
        print '{}: {} of {}'.format(cities_states[i][0], n, results)  # keep track of progress while running
        time.sleep(1)
        url = 'http://www.indeed.com/jobs?as_and=' + search + \
        '&as_phr=&as_any=&as_not=&as_ttl=&as_cmp=&jt=all&st=&salary=&radius=' \
        + str(miles) + '&l=' + cities_states[i][0] + '%2C+' + cities_states[i][1] + \
        '&fromage=any&limit=50&sort=date&start=' + str(n) + 'psf=advsrch'
        url_data = urllib2.urlopen(url).read()
        soup = BeautifulSoup(url_data)

        for row in soup('div', {'class' : 'result'}):
            data[counter] = {}
            if '<div class=\"sjcl\">' in str(row):    # only sponsored have this tag
                for sjcl in row('div', {'class': 'sjcl'}):  # special salary format for sponsored
                    for div in sjcl('div'):
                        if '<div>' in str(div): 
                            data[counter]['salary'] = div.get_text().strip()
                        else: 
                            data[counter]['salary'] = None
                data[counter]['post_type'] = 'sponsored'
            else:    
                for salary in row('td', {'class' : 'snip'}):  # salary for organic
                    if '<nobr>' in str(salary):
                        data[counter]['salary'] = salary.nobr.get_text().strip()
                    else: 
                        data[counter]['salary'] = None
                data[counter]['post_type'] = 'organic'

            # job title, company, location, and summary for organic and sponsored
            for job_title in row('a',{'data-tn-element' : 'jobTitle'}):
                data[counter]['job_title'] = append_result(job_title)
            for company in row('span', {'class' : 'company'}):
                data[counter]['company'] = append_result(company) 
            for location in row('span', {'class' : 'location'}):
                data[counter]['location'] = append_result(location)   
            for summary in row('span', {'class' : 'summary'}):
                data[counter]['summary'] = append_result(summary) 
            for date in row('span', {'class' : 'date'}):
                data[counter]['date_posted'] = append_result(date)
            data[counter]['city_searched'] = str(cities_states[i][0]) + ', ' + str(cities_states[i][1])

            counter += 1
```

```python
# create dataframe and save as csv
df = pd.DataFrame(data).T
df.to_csv('indeed_data_raw.csv', encoding= 'utf-8')
```

```python
# create salary multiplier based on fraction of year
def salary_increment(string):
    'assign numeric values based on salary increments'
    string = str(string)
    if 'year' in string:
        return 1
    elif 'month' in string:
        return 12
    elif 'week' in string:
        return 52
    elif 'day' in string:
        return 365
    elif 'hour' in string:
        return 52*40
    else:
        return None

dataz['salary_increment'] = dataz['salary'].apply(lambda x: salary_increment(x))

```

```python
def salary_min(x):
    'return lower bound of salary range'
    out = ''
    string = x[0]
    for char in string:
        if char.isnumeric() or char == '.':
            out += char
    return out

def salary_max(x):
    'return upper bound of salary range'
    out = ''
    if len(x) == 1:
        string = x[0]
        for char in string:
            if char.isnumeric() or char == '.':
                out += char
        return out
    else:
        string = x[1]
        for char in string:
            if char.isnumeric() or char == '.':
                out += char
        return out

# Define min and max of salary range for post
dataz['salary_split'] = dataz['salary'].str.split('-')   
dataz['salary_min_a'] = dataz['salary_split'].apply(lambda x: salary_min(x) if type(x) == list else x)
dataz['salary_max_a'] = dataz['salary_split'].apply(lambda x: salary_max(x) if type(x) == list else x)

for column in ['salary_min_a', 'salary_max_a', 'salary_increment']:
    dataz[column].fillna(0, inplace=True)
    dataz[column] = dataz[column].astype(float)

# calculate salary by averaging min and max
dataz['salary_min'] = dataz['salary_min_a'] * dataz['salary_increment']
dataz['salary_max'] = dataz['salary_max_a'] * dataz['salary_increment']
dataz['salary_yearly'] = (dataz['salary_min'] + dataz['salary_max']) / 2
```

```python
# remove salary calculation fields
dataz.drop(['salary', 'salary_increment', 'salary_split', 'salary_min_a',
           'salary_max_a', 'salary_min', 'salary_max'], axis=1, inplace=True)
```

```python
# remove capitalization discrepancies between words
dataz['summary'] = dataz['summary'].apply(lambda x: x.lower())
```

```python
# cleaned titles to more standardized format
# attempted to clean summary, but ran into unicode errors and couldn't resolve
def clean_title(x):
    'cleans string'
    words = x.split()
    new_title = ''
    for word in words:
        new_word = ''
        word = word.lower()
        if word in ['sr', 'sr.']:    # clean sr and jr
            word = 'senior'
        elif word in ['jr', 'jr.']:
            word = 'junior'
        for char in word:    # remove numeric and most non-alpha chars
            if char.isalpha() or char in ['+', '&']:
                new_word += char
            elif char.isnumeric():
                new_word += '' 
            else:
                new_word += ' '
        if len(new_word) > 1:  # ignore one letter words
            new_title += new_word + ' '
    return ' '.join(new_title.split()).title()  # removes excess spaces, add proper capitalization


dataz['job_title'] = dataz['job_title'].apply(lambda x: clean_title(x))
stats_model_data = dataz
```

```python
# reset index to avoid join problems with CountVectorizer
dataz.reset_index(inplace=True)
dataz.drop('index', axis= 1, inplace= True)

#export to CSV.  Will append test results to this dataset later
dataz.to_csv('indeed_clean.csv', encoding= 'utf-8')

```

#### Use location and title keywords as features

```python
train_data = stats_model_data[stats_model_data['salary_yearly'] > 0].drop(['salary_yearly', 
                                                                           'post_type', 
                                                                           'summary', 'company'], axis=1)
```

```python
# Create dummy variables for specific words in title
words = ['Senior', 'Junior', 'Engineer', 'Scientist', 'Analyst', 'Statistic'
         'Developer', 'Engineer', 'Manager', 'Director', 'Architect']
for word in words:
    train_data['title_' + word] = train_data['job_title'].apply(lambda x: 1 if word in x else 0 )
```

```python
# create x and y for location & title words model
x = train_data.drop(['high_salary', 'job_title'], axis =1)
x['intercept'] = 1
y = train_data['high_salary']

logit = LogisticRegressionCV(penalty='l2')
logit.fit(x, y)
logit.score(x, y)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

# Function to extract coefficients from model
def examine_coefficients(model, df):
    df = pd.DataFrame(
        { 'Coefficient' : model.coef_[0] , 'Feature' : df.columns}
    ).sort_values(by='Coefficient')
    return df[df.Coefficient !=0 ]

examine_coefficients(logit, x)
```

```python
# Create confusion Matrix
y_pred = logit.predict(x)

conmat = np.array(confusion_matrix(y, y_pred))

confusion = pd.DataFrame(conmat, index=['low_sal', 'high_sal'],
                         columns=['pred_low_sal', 'pred_high_sal'])
confusion
```

```python
# print classification report
cls_rep = classification_report(y, y_pred)
print(cls_rep)
```

__Accuracy__

78% of all predictions were correct

__Precision__ 

75% of pred_low_sal were correct

80% of pred_high_sal were correct

__Recall__

81% of low_sal were correctly identified

73% of high_sal were correctly identified

__Summary__

The model is more predictive than baseline, but could benefit from further iterations.  If this model always needed to be able to identify high salary positions (high 'high_sal' recall), and the user wasn't as worried about falsely identifying positions as high salary, the high salary confidence threshold could be lowered.  This would reduce the overall accuracy of the model.

```python
# create ROC and AUC
y_score = logit.decision_function(x)

FPR = dict()
TPR = dict()
ROC_AUC = dict() # {}

# For class 1, find the area under the curve
FPR[1], TPR[1], _ = roc_curve(y, y_score)
ROC_AUC[1] = auc(FPR[1], TPR[1])


# Plot of a ROC curve for class 1 (high_sal)
plt.figure(figsize=[11,9])
plt.plot(FPR[1], TPR[1], label='ROC curve (area = %0.2f)' % ROC_AUC[1], linewidth=4)
plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title('High Salary Predictions', fontsize=18)
plt.legend(loc="lower right")
plt.show()
```

### LogisticRegressionCV - L1 (Lasso)

```python
# Model using L1 regularization
logit = LogisticRegressionCV(penalty = 'l1', solver='liblinear')  # solver to liblinear to avoid errors on l1
logit.fit(x, y)
logit.score(x, y)
```

```python
examine_coefficients(logit, x)
```

Lasso method significantly reduced the complexity of the model, completely dropping many features from the model.

```python
# Create confusion Matrix
y_pred = logit.predict(x)

conmat = np.array(confusion_matrix(y, y_pred))

confusion = pd.DataFrame(conmat, index=['low_sal', 'high_sal'],
                         columns=['pred_low_sal', 'pred_high_sal'])
confusion
```

```python
# print classification report
cls_rep = classification_report(y, y_pred)
print(cls_rep)
```

__Summary__

The L1 regularization (Lasso) produced lower accuracy than L2 (Ridge).  The model maintained the low_sal recall of .81, but decreased its high_sal recall to .65.  It was more likely to incorrectly predict high_sal values as low_sal.  

However, by creating a more simplified model it may have reduced overfitting and prove to be superior in real world tests. 

```python
# create ROC and AUC
y_score = logit.decision_function(x)

FPR = dict()
TPR = dict()
ROC_AUC = dict() # {}

# For class 1, find the area under the curve
FPR[1], TPR[1], _ = roc_curve(y, y_score)
ROC_AUC[1] = auc(FPR[1], TPR[1])


# Plot of a ROC curve for class 1 (high_sal)
plt.figure(figsize=[11,9])
plt.plot(FPR[1], TPR[1], label='ROC curve (area = %0.2f)' % ROC_AUC[1], linewidth=4)
plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title('High Salary Predictions', fontsize=18)
plt.legend(loc="lower right")
plt.show()
```

## Use CountVectorization and GridSearch to optimize models

```python
# Instantiate Model
model = LogisticRegression()
model
```

```python
from sklearn.grid_search import GridSearchCV

# Choose Cs
Cs = [.01, .1, 1, 10, 100, 1000]

# Apply grid parameters
grid = GridSearchCV(estimator=model,
                    param_grid={'C' : Cs,
                               'fit_intercept' : [True, False],
                               'penalty' : ['l1', 'l2'],
                               })             
```

```python
# Create x and y
train_data = dataz[dataz['salary_yearly'] > 0].drop(['salary_yearly', 'job_title', 'post_type', 
                                                     'summary'], axis=1)
x = train_data.drop(['high_salary'], axis =1)
y = train_data['high_salary']

# fit to grid
grid.fit(x, y)
```

```python
# View best score parameters
best_model = grid.best_estimator_
print grid.best_score_, '\n'
print best_model
```

```python
examine_coefficients(best_model, x)
```

```python
# Create confusion Matrix
y_pred = best_model.predict(x)

conmat = np.array(confusion_matrix(y, y_pred))

confusion = pd.DataFrame(conmat, index=['low_sal', 'high_sal'],
                         columns=['pred_low_sal', 'pred_high_sal'])
confusion
```

```python
# print classification report
cls_rep = classification_report(y, y_pred)
print(cls_rep)
```

__Summary__

Despite the gridsearch producing a logistic regression model with a lower score than the manually selected location and title Ridge model (.74 vs .77), when put to the test it performed much better.

The gridsearch produced a Ridge model logistic regression that has superior accuracy, precision, and recall.
